### 3. Age

```python
print(df[df['Pclass'] == 1]['Age'].mean())
print(df[df['Pclass'] == 2]['Age'].mean())
print(df[df['Pclass'] == 3]['Age'].mean())
```

```python
def impute_age(cols):
    Age=cols[0]
    Pclass=cols[1]
    if pd.isnull(Age):
        if Pclass == 1:
            return np.round(38.233440860215055)
        elif Pclass == 2:
            return np.round(29.87763005780347)
        else:
            return np.round(25.14061971830986)
    else:
        return Age
```

```python
df['Age']=df[['Age','Pclass']].apply(impute_age,axis=1)
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.isnull(),cbar=False,yticklabels=False,cmap='viridis')
```

```python
df.dropna(inplace=True)
```

```python
df.reset_index(inplace=True)
```

```python
df.drop('index',axis=1,inplace=True)
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.isnull(),cbar=False,yticklabels=False,cmap='viridis')
```

```python
ticket = train['Ticket'].apply(lambda x:x.split())
```

```python
lst=[]
for x in range(len(ticket)):
    if len(ticket.iloc[x])==1:
        lst.append(ticket.iloc[x][0])
        continue
    elif len(ticket.iloc[x])==3:
        lst.append(ticket.iloc[x][2])
        continue
    elif len(ticket.iloc[x])==2:
        lst.append(ticket.iloc[x][1])
```

```python
len(lst)
```

```python
lst
```

```python
tick=np.array(lst)
```

```python
tick=pd.Series(tick)
```

```python
tick=pd.DataFrame(tick,columns=['Ticket No'])
```

```python
train2=train.copy()
```

```python
train_df=pd.concat([train2,tick],axis=1)
```

```python
len(train_df)
```

```python
train_df.drop('index',axis=1,inplace=True)
```

```python
train_df.sort_values('Ticket No',ascending=False)
```

```python
train_df['Ticket No'] = train_df['Ticket No'].apply(lambda x: float(x))
```

### Training the model

```python
X=train_df.drop(['Survived','Ticket No'],axis=1).values
y=train_df['Survived'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=101)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
model = Sequential()

model.add(Dense(9,activation='relu'))
model.add(Dense(6,activation='relu'))
model.add(Dense(3,activation='relu'))

model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(patience=25)
```

```python
model.fit(X_train,y_train,epochs=600,validation_data=(X_test,y_test),callbacks=[early_stop])
```

```python
plt.figure(figsize=(8,6))
ax=pd.DataFrame(model.history.history).plot()
ax.set_ylim(0.3,1)
```

```python
predictions = model.predict_classes(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
print('\n')
print(confusion_matrix(y_test,predictions))
```

### Taking into account complete dataset
